```python
import requests
from PIL import Image

import torch
from transformers import AutoProcessor, LlavaForConditionalGeneration

model_id = "HuggingFaceH4/vsft-llava-1.5-7b-hf-trl"

prompt = "A chat between a curious user and an artificial intelligence assistant. The assistant gives helpful, detailed, and polite answers to the user's questions. USER: <image>\nWhat are these?\nASSISTANT:"
image_file = '/kaggle/input/imagekk/ppl-positive-leadership-in-action-engagement-1600x900-1.jpg'
model = LlavaForConditionalGeneration.from_pretrained(
    model_id, 
    torch_dtype=torch.float16, 
    low_cpu_mem_usage=True,
).to(0)

processor = AutoProcessor.from_pretrained(model_id)


```

```python
import re
import os
def sort_files_numerically(folder_path):
    files = os.listdir(folder_path)
    files = [f for f in files if re.match(r'travis-\d+\.jpeg', f)]
    files.sort(key=lambda x: int(re.search(r'(\d+)', x).group(1)))
    return files
```

```python
from transformers import T5Tokenizer, T5ForConditionalGeneration

def handle_prompt(paragraph, prompt_type="summarize", prompt="", max_length=500, min_length=40, chunk_size=512):

    tokenizer = T5Tokenizer.from_pretrained('t5-small')
    text_model = T5ForConditionalGeneration.from_pretrained('t5-small')

    input_text = f"{prompt_type}: {prompt} " + paragraph
    inputs = tokenizer(input_text, return_tensors="pt", max_length=chunk_size, truncation=True, padding="longest")
    input_ids = inputs["input_ids"]
    attention_mask = inputs["attention_mask"]


    output_ids = text_model.generate(input_ids, max_length=max_length, min_length=min_length, length_penalty=2.0, num_beams=4, early_stopping=True, attention_mask=attention_mask)
    output_text = tokenizer.decode(output_ids[0], skip_special_tokens=True)

    return output_text


prompt3 = "brief description of the play leading up to the goal"

summary = handle_prompt(paragraph, prompt_type="summarize")
print("Summary:", summary)

answer = handle_prompt(paragraph, prompt_type="answer", prompt=prompt3)
print("Answer to prompt:", answer)
```
